Implimenting association rule

**Apriori Algorithm**

```python
data3=pd.read_csv('/content/book.csv')
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_itemsets=apriori(data3,min_support=.1,use_colnames=True)
frequent_itemsets
```

Rule 1

```python
rule1=association_rules(frequent_itemsets,metric='lift',min_threshold=.7)
rule1.sort_values('lift',ascending=False)
```

We now have over 100 association let us now reduce it by apply the conditions

```python
rule1[rule1.lift>1.8]
```

Now the no:of row's got reduced to 64

```python
rule1[['support','confidence','lift']].hist()
```

```python
plt.scatter(rule1['support'],rule1['confidence'])
plt.show()
```

```python
sns.scatterplot('support', 'confidence', data=rule1, hue='antecedents')
plt.legend(loc='upper left',bbox_to_anchor=(1,1))
plt.show()
```

RULE2

```python
rule2=association_rules(frequent_itemsets,metric='lift',min_threshold=2)
rule2.sort_values('lift',ascending=False)
```

The total no of rules now got reduced to 28

Rule 3

support =.15

```python
frequent_itemset2=apriori(data3,min_support=.15,use_colnames=True)
frequent_itemset2
```

```python
rule3=association_rules(frequent_itemset2,metric='lift',min_threshold=.8)
rule3.sort_values('lift',ascending=True)
```

```python
rule3.shape
```

With the min support =.15 and min_lift=.8 we are able to generate 11 item sets

```python
plt.figure(figsize=(10,10))
rule3[['lift','confidence','support']].hist()
```

```python
sns.scatterplot('support','confidence',data=rule3,hue='antecedents')
```
