**Building the apriori algorithm**

**Apriori algorithm**

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

**Rule1**
support =.1
confidence=.7

```python
frequent_items1=apriori(data,min_support=.1,use_colnames=True)
frequent_items1
```

```python
rule1=association_rules(frequent_items1,metric='confidence',min_threshold=.7)
rule1.sort_values('confidence',ascending=False)
```

By implimenting the above mentioned conditions we got 129 item sets

We can narrow it down by changing the parameters

```python
rule1[['support','confidence','lift']].hist()
```

```python
plt.scatter(rule1['support'],rule1['confidence'])
plt.show()
```

```python
sns.scatterplot('support', 'confidence', data=rule1, hue='antecedents')
plt.legend(loc='upper left',bbox_to_anchor=(1,1))
plt.show()
```

**Rule 2**
support=.25
confidence=.8


```python
frequent_items2=apriori(data,min_support=.25,use_colnames=True)
frequent_items2
```

```python
rule2=association_rules(frequent_items2,metric='confidence',min_threshold=.8)
rule2.sort_values('confidence',ascending=False)
```

With the above conditions we are only getting 5 frequent item set

```python
rule2[['support','confidence','lift']].hist()
```

```python
sns.scatterplot('support', 'confidence', data=rule1, hue='antecedents')
plt.legend(loc='upper left',bbox_to_anchor=(1,1))
plt.show()
```
