**Clustering Without using PCA**

Hierarchical clustering

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(15,15))
denedrogram=sch.dendrogram(sch.linkage(data1,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Wines')
plt.ylabel('Euclidean Distance')
```

By using ward method in hierarchichal clustering we can infer that the data set can be clustered into 3 groups(3'different colours)

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(data1)
```

```python
y_hc
```

```python
Hierarchical=pd.DataFrame(y_hc,columns=['Hierarchical cluster'])
```

```python
Hierarchical.head()
```

```python
plt.scatter(data1[y_hc==0,0],data1[y_hc==0,1],s=100,c='red',label='cluster1')
plt.scatter(data1[y_hc==1,0],data1[y_hc==1,1],s=100,c='blue',label='cluster2')
plt.scatter(data1[y_hc == 2, 0], data1[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
```

**Kmeans Clustering**

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit(data1)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
```

Based on the elbow plot we can select 3 as the  optimal no of clusters

```python
kcluster=KMeans(3,random_state=0)
kcluster.fit(data1)
```

```python
y_kmeams=kcluster.predict(data1)
```

```python
plt.scatter(data1[y_kmeams==0,0],data1[y_kmeams==0,1],s=100,c='red',label='cluster1')
plt.scatter(data1[y_kmeams==1,0],data1[y_kmeams==1,1],s=100,c='blue',label='cluster2')
plt.scatter(data1[y_kmeams == 2, 0], data1[y_kmeams == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.legend()
```

With out implimenting PCA we classified the data into 3 clusters using Kmeans also

**PCA**

```python
from sklearn.decomposition import PCA
pca_ob=PCA()
pca_values=pca_ob.fit_transform(data1)
```

```python
pca_values
```

```python
len(data.columns)
```

```python
pca=PCA(n_components=14)
pca_values=pca.fit_transform(data1)
```

```python
var=pca.explained_variance_ratio_
var
```

This show the amount of variance each of the features are holding

```python
#Cummulative variance
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

From the cumulative sum we can infer that first 11 columns holds  97.46 % information of the data set  

```python
data2=pd.DataFrame(pca_values,columns=['PCA1','PCA2','PCA3','PCA4','PCA5','PCA6','PCA7','PCA8','PCA9','PCA10','PCA11','PCA12','PCA13','PCA14',])
```
