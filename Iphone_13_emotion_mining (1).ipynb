**Cleaning the data**

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
tokenizer=RegexpTokenizer(r'\w+')
stopword=set(stopwords.words('english'))
ps=PorterStemmer()
```

```python
def get_clean_text(text):
  text=text.lower()     #Converting into lower case
  tokens=tokenizer.tokenize(text) #splitting into words
  new_token=[token for token in tokens if token not in stopword]   #Removing all stopwords
  stemmed_token=[ps.stem(token) for token in new_token]
  clean_text=" ".join(stemmed_token)
  return clean_text
```

```python
data2['clean_review']=data2['Reviews'].map(get_clean_text)
data2
```

**Now we have performed our cleaning operation .now using textblob we make emotion mining**

```python
from textblob import TextBlob as tb
```

```python
def sent_score(text):
  blob1=tb(text)
  score=blob1.sentiment[0]
  return score
```

```python
data2['sent_value']=data2['clean_review'].map(sent_score)
data2
```

Let us now create a function to classify the sentiment score into emotions

```python
def emotion(x):
  if x<-.5:
    y='worst'
  elif x<0:
    y='poor'
  elif x==0:
    y='average'
  elif x<.5:
    y='good'
  else:
    y='great'
  return y
```

```python
data2['Class']=data2['sent_value'].map(emotion)
data2
```

**N-gram**

BI-gram

```python
from sklearn.feature_extraction.text import CountVectorizer
def get_freq_words(corpus,gram,n=None):
  if gram==1:
    x=(1,1)
  elif gram==2:
    x=(2,2)
  elif gram==3:
    x=(3,3)
  else:
    x=(gram,gram)
  vec1=CountVectorizer(ngram_range=x,max_features=2000)
  bag_of_words=vec1.fit_transform(corpus)
  sum_words=bag_of_words.sum(axis=0)
  words_freq=[(word,sum_words[0,idx]) for word,idx in vec1.vocabulary_.items()]
  words_freq=sorted(words_freq,key=lambda X:X[1],reverse=True)
  return words_freq[:n]
```

```python
top_words=get_freq_words(data2['clean_review'],gram=2,n=200) #Top 200 frequent repeating words
top_word_df=pd.DataFrame(top_words)
top_word_df.columns=['Bi-gram','frequency']
top_word_df.head()
```

```python
top_word_bi=top_word_df.iloc[0:20,:]
plt.figure(figsize=(15,10))
plot=sns.barplot(x=top_word_bi['Bi-gram'],y=top_word_bi['frequency'])
plot.set_xticklabels(rotation=45,labels=top_word_bi['Bi-gram'])
```

```python
def split_into_words(i):
  return(i.split(' '))
```

**Creating countvectorizer matrix**

**Word count matrix**

```python
#Creating word count matrix for the eniter data set
comment_bow=CountVectorizer(analyzer=split_into_words).fit(data2.clean_review)
comment_bow_matrix=comment_bow.transform(data2.clean_review)
comment_bow_matrix.shape
```

```python
#bow for training dataset
train_bow_matrix=comment_bow.transform(comment_train.clean_review)
train_bow_matrix.shape
```

**The test accuracy of the model is 73%**
